### Transformations

```python
df_cal.display()
```

```python
df_cal = df_cal.withColumn("Year", year(col('Date')))\
         .withColumn("Month", month(col('Date')))
df_cal.display()
```

```python
df_cal.write.format("parquet")\
    .mode("append")\
    .option("path", "abfss://silver@projectdatalakeacc.dfs.core.windows.net/Calender")\
    .save()
```

```python
df_tet.display()
```

```python
df_tet.write.format("parquet")\
    .mode("append")\
    .save("abfss://silver@projectdatalakeacc.dfs.core.windows.net/Territories")
```

```python
df_cus = df_cus.withColumn("Full Name", concat_ws(" ", df_cus.Prefix, df_cus.FirstName, df_cus.LastName))\
               .withColumn("AnnualIncomeinRupees", regexp_replace(regexp_replace(df_cus.AnnualIncome, ',', ''), '\\$', '').cast("double"))
```

```python
df_cus = df_cus.withColumn("AnnualIncomeinRupees", df_cus.AnnualIncomeinRupees * 88.67)
```

```python
df_cus.printSchema()
```

```python
df_cus.write.format("parquet")\
    .mode("append")\
    .save("abfss://silver@projectdatalakeacc.dfs.core.windows.net/Customers")
```
